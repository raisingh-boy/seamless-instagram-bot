```python
# Download Wan 2.2 model
import torch
from diffusers import DiffusionPipeline
from huggingface_hub import snapshot_download

model_id = "Wan-AI/Wan2.2-T2V-A14B-Diffusers"

print('🔄 Loading model (this takes 2-3 min on first run)...')
pipe = DiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    variant="fp16"
)
pipe.to("cuda")
```

## Section 1 — Generate Reel

**Промты для CI & Emotional Intelligence:**

Копируй любой промт ниже в代码 ячейку и запускай.

```python
# === ПРОМТ ДЛЯ ГЕНЕРАЦИИ ===
# Выбери один из вариантов ниже, раскомментируй, убери # в начале

# Вариант 1 — CI partnering, slow motion
prompt = "Two dancers in contact improvisation, sharing weight, slow fluid movement, black background, soft spotlight, cinematic 4k, 24fps"

# Вариант 2 — Touch & trust
# prompt = "Close up of hands touching, contact improvisation dancers supporting each other, warm lighting, emotional, artistic"

# Вариант 3 — Consent through movement
# prompt = "Two dancers exploring consent through touch, one offers hand, other receives, gentle movement, studio, natural light"

# Вариант 4 — Lifts & trust
# prompt = "Contact improvisation lift, one dancer supports another in the air, trust, balance, contemporary dance, dark background"

# Вариант 5 — Spirals & flow
# prompt = "Dancers spinning in contact improvisation, connected through one point of contact, fluid motion, artistic lighting"

print(f'🎬 Generating: {prompt}')
print('⏳ This takes 2-3 minutes...')

# Generate 5 second video at 480p (free tier friendly)
video = pipe(
    prompt=prompt,
    num_frames=81,  # ~5 seconds at 16fps
    width=832,
    height=480,
    guidance_scale=5.0,
    num_inference_steps=30
)

```

```python
# Save and display video
import imageio
import numpy as np
from IPython.display import Video

frames = video.frames[0]
output_path = "/content/seamless_reel.mp4"

# Save as MP4
writer = imageio.get_writer(output_path, fps=16, codec='libx264', quality=8)
for frame in frames:
    writer.append_data(np.array(frame))
writer.close()

print(f'✅ Saved to {output_path}')
print(f'📹 Duration: ~{len(frames)/16:.1f}s')
print(f'📐 Resolution: 832x480')

# Display video in notebook
```
